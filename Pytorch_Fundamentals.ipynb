```python
# Reshaping, Stacking, Sqeezing, Unsqueezing tensors


x = torch.arange(1.,10.)
x
```

```python
x_reshape = x.reshape(1, 9)
x.shape , x_reshape.shape
```

```python
z = x.view(1,9)
z.shape
```

```python
x_stacked = torch.stack([x,x,x,x],dim=1)
```

```python
x_stacked
```

```python
x_stacked = torch.stack([x,x,x,x],dim=0)
x_stacked
```

```python
x_squeeze = torch.squeeze(x_reshape)

x_reshape.shape , x_squeeze.shape
```

```python
x_squeeze
```

```python
x_unsq = torch.unsqueeze(x_squeeze,dim=1)
x_unsq.shape
```

```python
# Reproducibility

RANDOM_SEED = 42

torch.manual_seed(RANDOM_SEED)
random_a = torch.rand(3,3)
torch.manual_seed(RANDOM_SEED)
random_b = torch.rand(3,3)

random_a == random_b
```
